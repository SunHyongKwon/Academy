### 지도에 마커 표시하기

```python
df = pd.read_excel('../Data/서울지역 대학교 위치.xlsx')
df.columns = ['학교명','위도','경도']
df.set_index('학교명',inplace=True)
df
```

```python
# 서울 지도 만들기
seoul_map = folium.Map(
    location = [37.55,126.98],
    zoom_start = 14
)

for lat , lng, name in zip(df.위도 ,df.경도 , df.index) :
    folium.Marker([lat, lng],popup=name).add_to(seoul_map)

seoul_map
```

---
### 지도에 원형 마커 표시하기

```python
# 대학교 위치정보를 CircleMarker로 표시하기
for lat , lng, name in zip(df.위도 ,df.경도 , df.index) :
    folium.CircleMarker(
        [lat, lng],
        radius=10, # 요걸 크기에 따라서 바꿔줄 수 있다.
        color='brown',
        fill=True,
        fill_color='coral',
        fill_opacity=0.7,
        popup=name).add_to(seoul_map)

seoul_map
```

### 지도 영역에 단계구분포(Choropleth Map) 표시하기
- 행정구역과 같이 지도상의 어떤 경계에 둘러싸인 영역에 색을 칠하거나 음영등의 정보를 시각화 하는 방법

```python
import json
```

```python
df=pd.read_excel('../Data/경기도인구데이터.xlsx')
df.set_index('구분',inplace=True)
# 다 int이다. 
df.info()
```

```python
# column은 string으로 바꿔줘야 된다. 
df.columns = df.columns.map(str)
df.columns
```

```python
df.head()
```

```python
geo_path = '../Data/경기도행정구역경계.json'
geo_data = json.load(open(geo_path,encoding='utf-8'))
geo_data
```

```python
df.head()
```

```python
# 경기도 지도 만들기
gyong_gi_do_map = folium.Map(
    location=[37.55,126.98],
    tiles='Stamen Terrain',
    zoom_start=9
)

# 출력할 연도 선택
year = '2017'

# chorpleth Class로 단계 구분 표시  
folium.Choropleth(
    geo_data=geo_data, # 지도 경계 데이터
    data=df[year], # 표시하려는 데이터
    columns=[df.index , df[year]], # 열 지정
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.3,
    threshold_scale=[10000, 100000, 300000, 500000, 700000],
    key_on='feature.properties.name'
).add_to(gyong_gi_do_map)

# 지도를 html 파일로 저장하기
gyong_gi_do_map.save('../Data/gyong_gi_do_map_pop_'+year+'.html')

gyong_gi_do_map
```
