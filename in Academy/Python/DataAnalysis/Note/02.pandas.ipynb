### 데이터 변경하기

```python
df['국어'] = 80
df
```

```python
df['국어'] = [10,20,30]
df['국어'][1] = 70
df
```

```python
# 행 추가
# 철수 학생 추가
df.loc['철수'] = [10,20,30,40,50]
df
```

```python
# 영희의 점수는 인아와 동일하다 데이터 추가하기
df.loc['영희'] = df.loc['인아']
df
```

```python
# 서준의 체육 점수를 80으로 변경하기
df.loc['서준','체육'] = 80

df.iloc[0,3] = 80
df
```

```python
# 서준의 음악과 체육 점수를 50점으로 변경하기
df.loc['서준',['음악','체육']] = 50
df.iloc[0,2:4] = 50
df
```

```python
df.loc['서준','음악':'체육'] = [100,50]
df.iloc[0,2:4] = [100,50]
df
```

```python
# T transpose 는 똑같다.
# df.T
df.transpose()
```

```python
df = pd.DataFrame({
    'c0':[1,2,3],
    'c1':[4,5,6],
    'c2':[7,8,9],
    'c3':[10,11,12],
    'c4':[13,14,15]
},index=['r0','r1','r2'])
df
```

```python
# 인덱스 재설정
new_index = ['r0','r1','r2','r3','r4']
ndf = df.reindex(new_index,fill_value=0)
ndf
```

```python
# 원래 있던 index는 column으로 넣고 없애고, 기본 index로 바꾼다.
ndf = df.reset_index()
ndf
```

```python
# 인덱스 정렬하기
df.sort_index()
df.sort_index(ascending=False)
```

```python
# c1을 기준으로 내림차순 정렬 후 바로 적용하기
df.sort_values(by='c1',ascending=False,inplace=True)
df
```
