```python
import pandas as pd
import numpy as np
```

```python
# numpy 난수를 이용하여 DataFrame 만둘기
df = pd.DataFrame(np.random.rand(6,4))

df
```

```python
# Column과 index 넣기
df.columns = ['A','B','C','D']
df.index = pd.date_range(start='20230126',periods=6)
df
```

```python
# 새로운 F열 생성과 값 입력
# nan을 만들 수 있는 것은 numpy 의 np.nan 밖에 없다. 
df['F'] = [1.0,np.nan,3.5,6.1,np.nan,7.0]
df
```

```python
# NaN이 하나라도 있는 Data행 삭제
df1 = df.copy()

df1.dropna(how='any',inplace=True,axis=0)
df1
```

```python
df2 = df.copy()

# NaN이 모든 행에 데이터로 있는 경우 삭제
df2.dropna(how='all',inplace=True)

df2
```

```python
# NaN 값을 특정 값으로 변경한다.
df3 = df.copy()

df3.fillna(value=5.0, inplace=True)
df3
```

```python
df.isnull()
```

```python
df4 = df.copy()
```

```python
# F열에서 NaN을 포함하고 있는 행 찾기
df4.loc[df4.F.isnull(),:]
```

```python
# index를 이용하여 행 제거
df4.drop('2023-01-27',inplace=True)
df4
```

```python
# F 열 삭제
df4.drop(columns='F',inplace=True)
df4
```

```python
# D 열 삭제
#del df4['D']
del df4['A']
df4
```

```python
# 열방향 합계 구하기
df.sum(axis='rows')

# 행방향 합계 구하기
df.sum(axis = 'columns')

df
```

```python
# 열방향 평균 구하기
print(df.mean(axis='rows'))

# 행방향 평균 구하기
# nan 스킵 안할시에는 nan이 있으면 계산 안한다. 
print(df.mean(axis = 'columns',skipna=False))

df
```

```python
data = [
    [1.4,np.nan],
    [7.1,-4.5],
    [np.nan,np.nan],
    [0.75,-1.3]
]

df = pd.DataFrame(data,columns=['one','two'],index=['a','b','c','d'])
df
```

```python
# df의 one컬럼의 NaN : one의 평균값으로 대체
# df의 two컬럼의 NaN : two의 최소값으로 대체
df.one.fillna(value=df.one.mean(),inplace=True)
df.two.fillna(value=df.two.min(),inplace=True)

df
```

### DataFrame Merging(병합)

```python
df1 = pd.DataFrame(
    {
        'key' : list('bbacaab'),
        'data1' : range(7)
    }
)

df1
```

```python
df2 = pd.DataFrame(
    {
        'key' : list('abd'),
        'data2' : range(3)
    }
)

df2
```

```python
# default : inner join 
pd.merge(df1,df2, on='key')

```

```python
# 모두 보이기 , outer join
pd.merge(df1,df2,on='key',how='outer')
```

```python
# df1 기준으로 합치기
pd.merge(df1,df2,on='key',how='left')
```

```python
# df2 기준으로 합치기
pd.merge(df1,df2,on='key',how='right')
```

### key name 이 다를 경우

```python
df1 = pd.DataFrame(
    {
        'lkey':list('bbacab'),
        'data1':range(6)
    }
)

df2 = pd.DataFrame(
    {
        'rkey':list('ababd'),
        'data1':range(5)
    }
)
```

```python
# default inner join
pd.merge(df1,df2,left_on='lkey',right_on='rkey')
```

```python
# outer join
pd.merge(df1,df2,left_on='lkey',right_on='rkey',how='outer')
```

```python
# left outer join
pd.merge(df1,df2,left_on='lkey',right_on='rkey',how='left')
```

```python
# right outer join
pd.merge(df1,df2,left_on='lkey',right_on='rkey',how='right')
```

### DataFrame의 Concatenation

```python
df1 = pd.DataFrame(
    np.arange(6).reshape(3,2),
    index=['a','b','c'],
    columns=['one','two']
)

df1
```

```python
df2 = pd.DataFrame(
    5 + np.arange(4).reshape(2,2),
    index=['a','c'],
    columns=['three','four']
)

df2
```

```python
# df1과 df2를 합치기
# column을 추가하면서 index가 같은 것 끼리 합쳐진다.
print(pd.concat([df1,df2],axis=1))
# index를 초기화 시킨다. 
print(pd.concat([df1,df2],axis=1,ignore_index=True))

# row 추가하면서 column name이  같은 것 끼리 합쳐진다.
print(pd.concat([df1,df2],axis=0))
# column name을 초기화 한다. 
pd.concat([df1,df2],axis=0,ignore_index=True)
```

### 데이터프레임 중복값 제거

```python
df = pd.DataFrame({
    'k1':['one'] * 3 + ['two'] * 4,
    'k2':[1,1,2,3,3,4,4]
})

df
```

```python
# 중복값 확인
df.duplicated(['k2'])
```

```python
# 중복값 제거
# 적용은 안함 할려면 inplace 사용
df.drop_duplicates(['k1','k2'])
```

```python
# 새로운 열 추가
df['v1'] = np.arange(7)

# k2 값들로 중복값 제거
df.drop_duplicates(['k2'])
```

```python
# 첫 데이터를 살리는 거 말고, 마지막 데이터를 살리고 싶을 때
df.drop_duplicates(['k1'],keep='last')
```

### Category

```python
df3 = pd.DataFrame(
    {
        'id':range(1,7),
        'raw_grade':['a','b','b','a','a','e']
    }
)

df3
```

```python
# category 자료형으로 변환하기
df3['grade'] = df3['raw_grade'].astype('category')
df3
```

```python
df3.grade
```

```python
df3.raw_grade
```

```python
# category를 이용하여 자료 변형
df3.grade.cat.categories = ['very good','good','very bad']
df3.grade.cat.codes.value_counts()
```

```python
df3.sort_values(by='grade')
```

### 데이터의 범위 정하기

```python
ages=[20,22,25,27,21,23,37,31,61,45,41,32]
bins=[18,25,35,60,100]
```

```python
cats = pd.cut(ages , bins)
cats
```

```python
# 범주 관련 Code 보기
cats.codes
```

```python
# 범주 관련 빈도수 보기
cats.value_counts()
```

```python
# Category 이름 정하기
group_names = ['Youth','YoungAdult','MiddleAged','Senior']
cat2 = pd.cut(ages,bins,labels=group_names)
cat2.value_counts()
```
