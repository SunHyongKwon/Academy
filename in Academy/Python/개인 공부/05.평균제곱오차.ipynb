```python
import pandas as pd
import numpy as np
```

```python
# 맞출 선 그리기
def ans(x):
    return 2*x + 5
```

```python
# 랜덤한 w 와 b 하기
w = 1.0
b = 1.0
```

```python
# 틀릭시 w와 b의 gradient 바꿀 편미분한 거 
def w_b_grd(x, y , w , b , gaesu):
    return gaesu * ((w * x + b - y) * w).mean() , gaesu * (w * x + b - y).mean()

def f(w,b,x):
    w_grad , b_grad = w_b_grd(x,w*x+b,w,b,0.7)
    return (w-w_grad) , (b-b_grad)
```

```python
# x 범위 잡기
x = np.arange(1,10)

# 처음 오차 구하기
oh_cha=np.mean(np.power((ans(x) - (w * x + b)),2))
```

```python
# 시각화 하면서 보기
import matplotlib.pyplot as plt

cnt = 0 

while oh_cha == 0:
    w_temp , b_temp =f(w,b,x)
    oh_cha_temp=np.mean(np.power((ans(x) - ((w - w_temp) * x + (b - b_temp)),2)))

    if oh_cha_temp < oh_cha:
        w = w_temp
        b = b_temp
        cnt += 1

plt.show()
```
