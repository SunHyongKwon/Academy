# 머신러닝(Machine Learning)이란?
수 많은 데이터를 학습시켜 거기에 있는 패턴을 찾아내고 이를 기반으로 데이터를 분류하거나 예측 <br><br>
- 일반 소프트웨어 <br>
    - 결정론적 소프트웨어이라고 한다. <br><br>
    - 언제나 똑같은 과정을 거쳐 똑같은 결과를 낸다. <br><br>
- 머신 러닝 기반 소프트웨어 <br>
    - 비결정론적 소프트웨어라고 한다. <br><br>
    - 데이터와 알고리즘에 따라 때로는 다른 결과를 추론해 낸다. <br><br><br><br>

- 추론이란 ? <br>
    - 이미 알고 있는 것으로 부터 논리적 결론을 도출하는 과정 <br><br>
    - y hat으로 표시한다.

---
### 머신러닝 개요
- 머신러닝 이란 인공지능 연구 과제 중의 하나로 인간의 뇌가 자연스럽게 수행하는 학습이나 능력을 컴퓨터로 구현하는 방법 <br><br>
    - 어느 정도 규모가 있는 샘플 데이터를 입력으로 넣어 분석 시킨다. <br><br>
    - 그리고 이 때 데이터에서 일정한 규칙을 찾아 낸다. <br><br>
    - 그리고 찾아낸 규칙을 기반으로 다른 데이터로 분류(Classification) 하거나 예측(Regression)하는 것 <br><br>



### 머신러닝의 종류
1) 지도학습 : Supervised Learning <br><br>
    - 학습시에 데이터와 답을 같이 학습한 후에 새로운 데이터로 값을 예측하는 것 <br><br>
    - 딥러닝과의 차이점 : 지도학습은 분석가가 관여하므로 결과가 달라짐 <br><br>
2) 비지도학습 : Unsupervised Learning <br><br>
    - 학습시 데이터만 학습한 후 다른 데이터의 규칙성을 찾는다. <br><br>
    - 본질적인 구조를 확인할 때 사용 : 군집화 <br><br>
    - 데이터가 너무 많거나 컬럼이 너무 많은 경우 줄이기 위해 사용 : 주성분분석 <br><br>
3) 강화학습 : Reinforcement Learning <br><br>
    - 부분적으로 답을 입력한 후 데이터를 기반으로 최적의 답을 찾아 낸다.<br><br>
    - 수많은 시행착오를 거쳐 답을 찾아 낸다. <br><br>
4) 딥러닝 : Deep Learning <br><br>
    - 엄밀히 따지면 지도학습의 한 부분이 딥러닝이다. <br><br>
    - 인공신경망을 확장시킨것이 딥러닝이다. <br><br>
    - 지도학습과 비지도학습과 관련이 있으나 특이성은 학습자료에 대한 인간의 개입을 받지 않는다. <br><br>

### 머신러닝 작업 효율
1) 데이터 수집 <br><br>
2) 데이터 가공<br><br>
    - na 있는지 , 타입이 맞는지 <br><br>
3) 데이터 학습<br><br>
    - 학습 방법 선택<br><br>
    - 매개 변수 조정 (hyper parameter) <br><br>
    - 모델 학습 <br><br>
4) 모델 평가 <br><br>
5) 예측력이 좋은 경우 다음 단계 진행하고 아니면 3번부터 다시 시작 <br><br>
6) 성공

### 머신러닝 응용 분야
#### 클래스 분류 (Classification) : 지도학습
- 특정 데이터에 레이블(정답)을 붙여 분류할 수 있다. <br><br>
- 스팸 메일 분류 , 필기인식 , 증권 사기 방지 등 <br><br>

#### 클러스터링 (Clustering) : 비지도학습
- 값의 유사성을 기반으로 데이터를 여러 그룹으로 나눌 수 있다. <br><br>
- 사용자의 취향을 그룹으로 묶어 사용자 취향에 맞는 광고를 제공 , 날짜별로 지역별로 많이 팔리는 물품 분류 <br><br>

#### 추천 (Recommendation) : 지도학습
- 특정 데이터를 기반으로 다른 데이터를 추천하는 것 이다. <br><br>
- 인터넷 서점에서 구매한 책들을 기반으로 다른 책을 추천 <br><br>

#### 회귀 (Regression) : 지도학습
- 과거의 데이터를 기반으로 미래를 예측하는 것 이다. <br><br>
- 판매 예측 , 주가 변동 <br><br>

#### 차원 축소 (Dimensionality Reduction) : 지도학습
- 데이터의 특성을 유지하면서 데이터의 양을 줄이는 것 <br><br>
- 데이터를 시각화 하거나 구조를 추출해서 용량을 줄여 계산을 빠르게 하거나 메모리를 절약할 대 사용 <br><br>


### 프로젝트 과정
머신러닝 프로젝트는 크게 문제 정의 , 데이터 획득 , 모델 구현 , 검증 , 테스트의 단계로 구현 된다. 

#### 문제정의
- 프로젝트의 목표를 정의하고 그에 따른 계획을 세우는 단계 <br><br>
- 대략적으로 어떤 데이터를 수집해서 어떤 머신러닝 알고리즘으로 문제를 해결할지 결정하는 단계이다. <br><br>

#### 데이터 획득
- 머신러닝 모델 학습에 필요한 데이터를 수집하는 단계 <br><br>
- 성공적인 머신러닝 프로젝트는 뛰어난 머신러닝 실력보다도 얼마나 충분한 데이터를 획득했는냐에 의해 결정된다. <br><br>
- 획득한 데이터는 학습데이터(train) , 검증데이터(valid) , 테스트데이터(test)로 구분된다. <br><br>
- 전체 데이터 중에 20%가 테스트 데이터, 나머지 80%중에서 80%정도가 학습데이터 이고 20%가 검증 데이터이다. <br><br>
- 이렇게 해야 일반화 성능을 볼 수 있다. <br><br>

#### 모델 구현
- 학습 데이터를 기반으로한 한개 이상의 머신러닝 모델을 구현 <br><br>

#### 검증
- 검증 데이터를 사용해 구현된 머신러닝 모델들의 성능을 검증 <br><br>
- 최종적으로 검증 결과가 가장 높은 모델 하나를 선택 <br><br>

#### 테스트
- 검증, 결과를 통해 선택된 모델의 성능을 테스트 데이터로 측정 (일반화 성능을 측정하는 것이다.) <br><br>
