### Numpy를 Image로 변환

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np

fruits = np.load('../Data/fruits.npy')
fruits.shape
```

```python
# Image로 저장하기
from PIL import Image
```

```python
for i in range(300):
    image = Image.fromarray(fruits[i]).convert('L') # 흑백톤
    image.save('../Data/fruits/fruit_%03d.jpg'%i,'JPEG')
```

---
### Image를 읽어서 Numpy로 변환하기

```python
import glob
```

```python
numberOfData = 300
imgSize = 100
```

```python
# 배열 초기화
fruitsData = np.zeros(numberOfData * imgSize * imgSize , dtype=np.int32).reshape(numberOfData,imgSize,imgSize)
```

```python
i = 0
for file in sorted(glob.glob('../Data/fruits/*.jpg')):
    img = np.array(Image.open(file),dtype=np.int32)
    fruitsData[i,:,:] = img
    i+=1

fruitsData.shape
```

```python
import matplotlib.pyplot as plt

plt.imshow(fruitsData[0],cmap='gray_r')

plt.show()
```

```python
# 저장하기 : npy로 , 뒤에 확장자명 안적어줘도 된다. 
np.save('../Data/fruitsData',fruitsData)
```
