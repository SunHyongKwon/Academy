```python
import urllib.request as req

# 이 이름의 파일을 가져오겠다. 
fileName = "mov01.jpg"

url = "https://zeushahn.github.io/Test/images/" + fileName

# 저장할 이름
saveName = "../Data/" + fileName

req.urlretrieve(url, saveName)
print("OK")
```

```python

import urllib.request as req

fileNames = ["mov02.jpg","mov03.jpg","mov04.jpg"]

for fileName in fileNames:
    url = "https://zeushahn.github.io/Test/images/" + fileName

    saveName = "../Data/" + fileName

    req.urlretrieve(url, saveName)
    print("OK")
```

### 기상청 
- 기상청의 데이터를 rss 라고 한다. <br><br>
- url :http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp <br><br>
- stnId -> 지역번호를 의미

|지역번호|지역|
|---|---|
|전국|108|
|서울/경기도/인천|109|
|강원도|105|
|전라북도|146|
|전라남도|156|
|제주특별자치도|184

```python
# 서울 / 경기도 / 인천 기상청 데이터 가져오기
import urllib.request as req

url = "https://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=109"
data = req.urlopen(url).read().decode("utf-8")
print(type(data))
print(data)
```
