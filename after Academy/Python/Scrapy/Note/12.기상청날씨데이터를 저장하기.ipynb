```python
import requests
from bs4 import BeautifulSoup
```

```python
req = requests.get('http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108')
html = req.text
```

```python
soup = BeautifulSoup(html,'html.parser')
```

```python
for i in soup.find_all('location'):
    print(i.find('city'))
```

```python
# 서울로 테스트
seoul = soup.find('location')

# tmef : 날짜 , wf : 날짜 , tmn : 최저기온 , tmx : 최고기온
for i in seoul.find_all('data'):
    print(i.find('tmef').text)
    print(i.find('wf').text)
    print(i.find('tmn').text)
    print(i.find('tmx').text)
```

```python
# 도시별 날씨 데이터 하기
weather = {}

for city in soup.find_all('location'):
    weather[city.find('city').text] = []
    for i in city.find_all('data'):
        temp = []
        temp.append(i.find('tmef').text)
        temp.append(i.find('wf').text)
        temp.append(i.find('tmn').text)
        temp.append(i.find('tmx').text)
        weather[city.find('city').text].append(temp)
```

```python
weather
```

```python
# csv file로 저장하기
for i in weather:
    for j in weather[i]:
        print(i,j[0],j[1],j[2],j[3])
    print("-"*20)
```

```python
# 리스트로 만들기
tempList = []
for i in weather:
    for j in weather[i]:
        tempList.append((i,j[0],j[1],j[2],j[3]))

tempList
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(data=tempList, columns=['지역','날짜','날씨','최저온도','최고온도'])
df
```

```python
# file로 저장
df.to_csv('../Data/weather.csv',index=False,encoding='utf-8')
```

### DB에 저장하기

#### SQLite

```python
import sqlite3
```

```python
dbPath = '../Data/weather.sqlite'
conn = sqlite3.connect(dbPath)

# 테이블을 생성하고 데이터 넣기
curs = conn.cursor()
# 여러줄
curs.executescript(
    """
    DROP TABLE IF EXISTS area_weather;

    CREATE TABLE area_weather (
        item_id INTEGER PRIMARY KEY AUTOINCREMENT,
        area TEXT,
        date TEXT,
        weather TEXT,
        max_temp INTEGER,
        min_temp INTEGER
    );
    """
)

# 위의 작업을 데이터베이스에 반영하기
conn.commit()
```

```python
forecast.head()
area = forecast['지역']
date = forecast['날짜']
weather = forecast['날씨']
max_temp = forecast['최고온도']
min_temp= forecast['최저온도']
```

```python
data=list(zip(area,date,weather,max_temp,min_temp))

curs = conn.executemany('insert into area_weather (area,date,weather,max_temp,min_temp) values (?,?,?,?,?)',data)

conn.commit()
```

```python
curs.execute('select * from area_weather')
items = curs.fetchall()
items
```
