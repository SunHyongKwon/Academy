---
### 특성(Feature) 컬럼 생성

```python
# Embarked에 따른 컬럼을 생성
# Q , S -> UK , C -> Austria
#True if train.Embarked == 'C' else False
train['Nationality_AT']= train.Embarked == 'C'
train['Nationality_UK']= train.Embarked != 'C'

train[['Embarked','Nationality_AT','Nationality_UK']].head()
```

```python
train['Nationality']=['Austria' if i == 'C' else 'England' for i in train.Embarked]
train['Nationality']=['England' if i else 'Austria' for i in train.Nationality_UK]
train['Nationality']=['Austria' if i else 'England' for i in train.Nationality_AT]

train[['Embarked','Nationality_AT','Nationality_UK','Nationality']].head()
```

```python
# 다른 방법
train.loc[train.Embarked == 'C','Nationality2'] = 'Austria'
train.loc[train.Embarked != 'C','Nationality2'] = 'England'
train.head()
```

```python
train = train.iloc[:,:-1]
```

```python
# fare 에 대한 분류 컬럼을 만들려고 한다. 

# 분위 수 확인 방법 1
print(train.Fare.describe())

# 분위 수 확인 방법 2
print(train.Fare.quantile(.25),train.Fare.quantile(.50),train.Fare.quantile(.75),train.Fare.quantile(1))
```

```python
# 2분위수와 3분위수의 근사 정수치를 선택
train['LowFare'] = train['Fare'] < 15
train['MediumFare'] = (train['Fare'] >= 15) & (train['Fare'] < 30)
train['HighFare'] = train['Fare'] >= 30

train[['Fare','LowFare','MediumFare','HighFare']].head()
```

```python
train.loc[train.LowFare,'FareType'] = 'Low'
train.loc[train.MediumFare,'FareType'] = 'Medium'
train.loc[train.HighFare,'FareType'] = 'High'

train[['FareType','LowFare','MediumFare','HighFare']].head(10)
```

```python
# 파이썬의 함수를 이용하여 처리하기
def check_fare_type(fare):
    if fare < 15:
        return 'Low'
    elif fare >= 15 and fare < 30:
        return 'Medium'
    else:
        return 'High'
```

```python
train.head()
```

```python
# apply에 들어갈 함수는 리스트의 형태가 아닌 원소의 형태로 매개변수가 들어가게끔 해야 된다.

train['FareType2'] = train['Fare'].apply(check_fare_type)
train[['Fare','FareType','FareType2']].head(10)
```

```python
# Age의 NaN을 Age의 평균값으로 대체하기 (단 소수점 첫째짜리 까지 반올림 하기)
train.loc[train.Age.isnull(),'Age'] = round(train.Age.mean(),1)
```
