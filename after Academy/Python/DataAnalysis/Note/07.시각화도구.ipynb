```python
def makeLine(sub_plot, data, name, color): 
    sub_plot.plot(
        data.columns,
        data.loc[name],
        marker = 'o',
        markerfacecolor = color,
        markersize = 5,
        color = color,
        linewidth = 2,
        label= name
    )

    sub_plot.set_ylim(8000,70000)
    sub_plot.set_xlabel('년도',size= 5)
    sub_plot.set_xticklabels(df_4.columns,rotation=90,size=6)

    sub_plot.set_ylabel('인구수',size= 5)

    sub_plot.legend(loc='best')
    sub_plot.set_title(f'서울 -> {name}',size = 20)

def makeSubPlot(fig, data , width , height, place , color , name):
    sub =fig.add_subplot(width, height, place)
    
    makeLine(sub,data,name,color)
```

```python
plt.style.use('ggplot')

fig = plt.figure(figsize=(15,15))

colors = ['yellowgreen','salmon','peachpuff','maroon']
regions = ['충청남도','경상북도','강원도','전라북도']

for i in range(1,5):
    makeSubPlot(fig,df_4,2,2,i,colors[i-1],regions[i-1])

plt.show()
```

```python
df = pd.read_excel('../Data/남북한발전전력량.xlsx',convert_float=True)
df = df.fillna(method='ffill')
```

```python
df_north=df.iloc[5:,:]
df_north=df_north.iloc[:,1:]
df_north.set_index('발전 전력별',inplace=True)
```

```python
# 돌리기
df_north=df_north.T
```

```python
df_north.rename({'합계':'총발전량'},axis=1,inplace=True)
```

```python
# 전년도 비교 하기 위한 새컬럼 만들기
# shift를 이용하여 한칸씩 밀 수 있다.
df_north['전년도 총발전량'] = df_north['총발전량'].shift(1)
df_north['총발전량 차이'] = df_north.총발전량 - df_north['전년도 총발전량']
df_north
```

```python
df_north['증감율']= (df_north['총발전량'] / df_north['전년도 총발전량'] - 1) * 100
df_north
```

```python
#그래프 그리기
plt.style.use('ggplot')

fig = plt.figure(figsize=(15,8))

ax1 = fig.add_subplot(1,1,1)

df_north[['수력','화력']].plot(
    kind = 'bar',
    figsize=(15,8),
    stacked = True,
    ax=ax1
)

ax2 = ax1.twinx()
ax2.plot(
    df_north.증감율,
    ls='--',
    marker = 'o',
    markersize = 10,
    color = 'green',
    label = '전년대비 증감율(%)'
    )
    
ax1.set_ylim(0,500)
ax2.set_ylim(-50,50)

ax1.set_xlabel('연도',size = 20)
ax1.set_ylabel('발전량(억 Kwh)')
ax2.set_ylabel('전년대비 증감율(%)')

plt.title('북한 전력 발전량(1990 ~ 2016)',size=20)

ax1.legend(loc='upper left',fontsize = 20)
#위쪽은 upper, 아래쪽은 lower

plt.show()
```
