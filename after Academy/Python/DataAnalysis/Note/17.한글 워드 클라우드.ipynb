```python
import nltk
from konlpy.corpus import kolaw
from konlpy.tag import Okt
import matplotlib.pyplot as plt
from wordcloud import WordCloud

t = Okt()
```

```python
ko_conn_text = kolaw.open('constitution.txt').read()
ko_conn_text
```

```python
token_ko = t.nouns(ko_conn_text)
token_ko
```

```python
# 불용어 처리
stop_words = ['제','월','일','그','이','바','및','기타','입','정','수','관','때','위','안','장']
token_ko = [each_word for each_word in token_ko if each_word not in stop_words]
token_ko
```

```python
# 상위 50개 단어 확인
ko = nltk.Text(token_ko)
ko.vocab().most_common(50)
```

```python
# 상위 500개만 wordcloud로 그리기
data =ko.vocab().most_common(500)
data
```

```python
tmp_data = dict(data)
wordcolud = WordCloud(
    background_color='white',
    font_path='AppleGothic',
    relative_scaling=0.2
).generate_from_frequencies(tmp_data)
```

```python
plt.figure(figsize=(12,10))

plt.imshow(wordcolud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
