```python
def calcInterest(money, interest1 , interest2) :
    large = money * 0.9
    small = money * 0.1

    geon = large * interest1 / 100 / 12
    sin = small * interest2 / 100 / 12

    return geon + sin
```

```python
import numpy as np 
for i in range(10,23):
    money = i / 10 * 10**8
    interest = calcInterest(i / 10 * 10**8, 1.2 , 4.17)

    print(f"전체 돈 : {format(round(money), ',')}원")
    print(f"내야할 이자 : {format(round(interest), ',')}원")
    print('-'*100)
```
